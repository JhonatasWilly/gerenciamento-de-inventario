```python
                                       # Bibliotecas necessárias

# Importa a biblioteca csv para possibilitar geração de arquivos no formato csv
import csv

# Importa a biblioteca os para manipulação de arquivos
import os

# Importa datetime para capturar data e hora da venda
from datetime import datetime

produtos = {                            # Dicionario de Produtos
    1: {
        "nome": "Smartphone XYZ", "preço": 899.99, "quantidade": 30, "categoria": "Eletrônicos", "ativo": True,
    },
    2: {
        "nome": "Notebook Ultra", "preço": 1599.99, "quantidade": 15, "categoria": "Eletrônicos", "ativo": True,
    },
    3: {
        "nome": "TV 4K 55''", "preço": 2199.99, "quantidade": 10, "categoria": "Eletrônicos", "ativo": True,
    },
    4: {
        "nome": "Fone de Ouvido Bluetooth", "preço": 199.99, "quantidade": 50, "categoria": "Acessórios", "ativo": True,
    },
    5: {
        "nome": "Mochila Executiva", "preço": 89.99, "quantidade": 25, "categoria": "Acessórios", "ativo": True,
    },
    6: {
        "nome": "Livro - Data Science", "preço": 39.99, "quantidade": 80, "categoria": "Livros", "ativo": True,
    },
    7: {
        "nome": "Caneta Esferográfica", "preço": 1.99, "quantidade": 200, "categoria": "Papelaria", "ativo": True,
    },
    8: {
        "nome": "Caderno Universitário", "preço": 14.99, "quantidade": 100, "categoria": "Papelaria", "ativo": True,
    },
    9: {
        "nome": "Cafeteira Elétrica", "preço": 129.99, "quantidade": 40, "categoria": "Eletrodomésticos", "ativo": True,
    },
    10: {
        "nome": "Liquidificador Turbo", "preço": 249.99, "quantidade": 20, "categoria": "Eletrodomésticos", "ativo": True,
    }
}

# Inicializa um dicionário que armazenrá as vendas já realizadas e as futuras vendas
vendas = {}

# Inicializa um dicionário que armazenará as IDs das vendas realizadas
relatorios = {}

def verificar_negativo(valor):          # Função Verifica se o valor é menor ou igual a zero.
    if valor <= 0 or valor > max(produtos.keys()):
        raise ValueError

def verificar_vazio(valor):             # Função Verifica se a variavel esta vazia.
    if valor == "":
        raise ValueError

def gerar_novo_id(produto_ID):          # Função Gerar ID automáticamente no dicionario de produtos
    if produto_ID:
        return max(ID for ID in produto_ID.keys()) + 1
    else:
        return 1

def novo_id_relatorio_vendas(dicionario):
    if dicionario:
        return max(dicionario.keys()) + 1
    else:
        return 1

def verificar_valores_validos_compras(listar):
    if listar < 1 or listar > 3:
        raise ValueError("Erro. Você deve informar apenas 1, 2 ou 3.")

def cadastrar_produto():                # 1 - função para cadastrar produtos.
    print("\n--- Cadastrar novo Produto ---")

    identificador = gerar_novo_id(produtos)

    # validar nome.
    while True:
        try:
            nome = input("Digite o nome do produto: ")
            verificar_vazio(nome)
            break
        except ValueError:
            print("Erro: O nome é obrigatório e não pode estar vazio")

    # validar preço.
    while True:
        try:
            preco = float(input("Digite o preço do produto: "))
            verificar_negativo(preco)
            break
        except ValueError:
            print("Erro: Valor inválido. O preço deve ser informado e ser um número positivo maior que 0 e não pode conter letras.")

    # validar quantidade.
    while True:
        try:
            quantidade = int(input("Digite a quantidade em estoque disponível: "))
            verificar_negativo(quantidade)
            break
        except ValueError:
            print("Erro: Valor inválido. A quantidade em estoque deve ser informada e ser um número positivo maior que 0 e não pode conter letras.")

    # validar categoria.
    while True:
        categoria = input("Digite a categoria do produto: ")
        try:
            verificar_vazio(categoria)
            break
        except ValueError:
            print("Erro: A categoria é obrigatória e não pode estar vazia")

    # Descrição por ser opcional não tem validação.
    descricao = input("Digite a descrição do produto (ou pressione enter se não quiser adicionar): ")

    # adiciona o novo produto ao dicionario de produtos.
    produtos[identificador] = {
        "nome": nome, "preço": preco, "quantidade": quantidade, "categoria": categoria, "ativo": True
    }

    # caso tenha descrição é adicionada ao produto
    if descricao:
        produtos[identificador]["descrição"] = descricao

    return f"{nome} cadastrado com sucesso!"

def listar_produtos(lista_produtos):    # 2 - função Listar todos os Produtos
    print("\n--- Lista de Produtos ---\n")
    for chave, produto in lista_produtos.items():
        if produto["ativo"] == True:
            print(chave, produto["nome"])
    print('\n')

def atualizar_produto():                # 3 - função para atualizar produtos.
    print("\n--- Atualização de Produto ---")
    while True:
        try:
            listar = input("Se desejar Digite 0 para listar os Produtos disponiveis: ")
            if listar == "0":
                listar_produtos(produtos)

            identificador = int(input("\nDigite o número do produto que deseja atualizar: "))
            if produtos[identificador]["ativo"] == False:
                return "\nErro: Produto não disponivel"
            verificar_negativo(identificador)
            if identificador in produtos.keys():
                for chave, valor in produtos[identificador].items():
                    print(f"{chave}: {valor}")
                break
        except ValueError:
            print('\nErro: Número do produto inválido. Por favor, insira um número válido para atualizar o produto desejado.')

    # Valida a Atualização do Preço.
    while True:
        try:
            preco = float(input("\nDigite o novo preço do produto: "))
            verificar_negativo(preco)
            break
        except ValueError:
            print("\nErro: Valor inválido. O novo preço deve ser um número positivo maior que 0 e não pode conter letras ")

    # Valida a Atualização da Quantidade em estoque.
    while True:
        try:
            quantidade = int(input("\nDigite a nova quantidade em estoque disponível: "))
            verificar_negativo(quantidade)
            return f"\n{produtos[identificador]['nome']} atualizado com sucesso!"
        except ValueError:
            print("\nErro: Valor inválido. A quantidade em estoque deve ser um número positivo maior que 0 e não pode conter letras ")

def excluir_produto():                  # 4 - função para excluir produtos.
    print("\n--- Exclusão de Produto ---")
    while True:
        try:
            listar = input("Digite 0 para visualizar os produtos disponíveis para exclusão ou 1 para sair: ")
            if listar == "0":
                listar_produtos(produtos)
            if listar == "1":
                break

            identificador = int(input("Digite o identificador do produto que deseja excluir: "))
            verificar_negativo(identificador)

            if identificador in produtos:
                produtos[identificador]["ativo"] = False
                return f"\n{produtos[identificador]['nome']} excluído(a) com sucesso!"

        except ValueError:
            print("Produto não encontrado! Digite um identificador válido!")

def realizar_venda():                   # 5 - função para Realização de Venda.
    print("\n--- Realização de Venda ---")

    carrinho = {} # dicionário para armazenar a compra de diferentes produtos de uma vez só

    while True:
        try:
            print('1. Visualizar Produtos disponíveis para vendas')
            print('2. Prosseguir com a venda')
            print('3. Finalizar Venda')
            print('4. Voltar ao Menu Principal')
            listar = int(input('Digite uma Opção: '))

            if listar == 1:
                listar_produtos(produtos)

            elif listar == 2:
               cont_erros = 0
               while True:
                  try:

                    if cont_erros == 3:
                        break

                    identificador = int(input("\nDigite o identificador do produto que deseja realizar a venda: "))
                    verificar_negativo(identificador)

                    if produtos[identificador]["ativo"] == False:
                        print("\nERRO: Identificador não corresponde a nenhum produto.\n")
                        break

                    if identificador in produtos:
                      quantidade = int(input("\nDigite a quantidade que deseja vender: "))

                      if quantidade <= produtos[identificador]["quantidade"] and quantidade > 0:

                          if identificador in carrinho:
                            carrinho[identificador] += quantidade
                          else:
                            carrinho[identificador] = quantidade

                          print(f"\n{quantidade} unidade(s) do produto {produtos[identificador]['nome']} adicionada(s) ao carrinho.\n")
                          break

                      elif quantidade == 0:
                        print('Erro. A quantidade deve ser maior que 0 (zero)\n.')
                        break

                      else:
                        print(f'\nQuantidade insuficiente no estoque.')
                        print(f"{produtos[identificador]['nome']}, temos: {produtos[identificador]['quantidade']} Disponíveis.\n")
                        break

                  except ValueError:
                    cont_erros += 1
                    print("Erro. Opção invalída!\n")

            elif listar == 3:
                if not carrinho:
                    print("O carrinho está vazio. Adicione produtos antes de finalizar a venda.\n")
                    continue

                total_venda = 0
                for identificador, quantidade in carrinho.items():
                    produtos[identificador]["quantidade"] -= quantidade
                    valor_venda = quantidade * produtos[identificador]["preço"]

                    # bloco que cria um novo dicionário com as infos da nova venda
                    chave_vendas = gerar_novo_id(vendas)

                    vendas[chave_vendas] = {
                        "nome": produtos[identificador]["nome"],
                        "preco": produtos[identificador]["preço"],
                        "quantidade_venda": quantidade,
                        "total_venda": valor_venda,
                        "categoria": produtos[identificador]["categoria"]
                        }
                    total_venda += valor_venda
                return f"\nVenda realizada com sucesso! Total: R${total_venda:.2f}"

            elif listar == 4:
              return

            else:
              print("Erro. Opção invalída!\n")

        except ValueError:
            print("Erro. Digite uma opção valída!\n")

def menu_principal():                   # Função Menu principal.
    while True:
        print("\n--- Menu Principal ---")
        print("1. Cadastro de produtos")
        print("2. Lista de Produtos")
        print("3. Atualização de produtos")
        print("4. Exclusão de produtos")
        print("5. Realização de vendas")
        print("6. Relatório de vendas")
        print("7. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            print(cadastrar_produto())
        elif opcao == "2":
            listar_produtos(produtos)
        elif opcao == "3":
            print(atualizar_produto())
        elif opcao == "4":
            print(excluir_produto())
        elif opcao == "5":
            print(realizar_venda())
        elif opcao == "6":
            relatorio_vendas()
        elif opcao == "7":
            print("Saindo do sistema...")
            break
        else:
            print("Opção inválida, por favor escolha novamente.")


def verificar_e_incluir_cabecalho(nome_arquivo, cabecalho):
    # Verifica se o arquivo existe ou se está vazio, e adiciona o cabeçalho se necessário
    arquivo_existe = os.path.exists(nome_arquivo)

    if not arquivo_existe or os.path.getsize(nome_arquivo) == 0:
        with open(nome_arquivo, 'w', encoding='utf-8', newline='') as arquivo:
            escritor = csv.writer(arquivo, delimiter=';', lineterminator='\n')
            escritor.writerow(cabecalho)
        print(f"Cabeçalho adicionado ao arquivo: {nome_arquivo}")
    else:
        print(f"O arquivo {nome_arquivo} já possui um cabeçalho ou não está vazio.")

def relatorio_vendas(): # função para criar relatórios de vendas de produtos.

    print("\n--- Relatório de Vendas ---")

    nome_arquivo = 'relatorio_vendas.csv'
    cabecalho = ['Nome', 'Qtd_Vendida', 'Preco_Un', 'Total_Venda', 'Data_Hora']
    verificar_e_incluir_cabecalho(nome_arquivo, cabecalho)

    try:
      # Definindo as regras do nosso CSV:
      # Uso do método "with" abre o arquivo para imputar informaçõese e fecha ele automaticamente, sem necessidade do comando "close()"
      # Ele será escrito no arquivo apontado pela variável 'arquivo'
      # O arquivo será aberto em modo de escrita ('w')
      # O arquivo será codificado em UTF-8 ('utf-8'), código que aceita linguagem latina
      with open(nome_arquivo, mode='a', encoding='utf-8', newline='') as arquivo:

        # Seus elementos serão delimitados (delimiter) pelo símbolo ';'
        # Suas linhas serão encerradas (lineterminator) por uma quebra de linha
        escritor = csv.writer(arquivo, delimiter='|', lineterminator='\n')

        # cria um cabeçalho com os "tópicos" contidos no relatório
        # escritor.writerow(['Nome', 'Qtd_Vendida', 'Preco_Un', 'Total_Venda', 'Data_Hora'])

        # o laço faz o programa percorrer o dicionário "relatório", onde cada chave representa um ID de venda, e venda é o dicionário com os detalhes daquela venda.
        for venda in vendas.values():
          Nome = venda["nome"]
          Qtd_Vendida = int(venda["quantidade_venda"])
          Preco_Un = venda["preco"]
          Total_Venda = float(Qtd_Vendida * Preco_Un)
          Data_Hora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
          escritor.writerow([Nome, Qtd_Vendida, Preco_Un, Total_Venda, Data_Hora])

      print(f"Relatório de vendas gerado com sucesso: {nome_arquivo}")
    except Exception as e:
      print(f"Ocorreu um erro ao gerar o relatório de vendas: {e}")

def carregar_vendas_anteriores():
    nome_arquivo = 'relatorio_vendas.csv'
    try:
        with open(nome_arquivo, mode='r', encoding='utf-8') as arquivo:
            leitor = csv.reader(arquivo, delimiter='|')
            next(leitor)

            for linha in leitor:
              if linha:
                Nome, Qtd_Vendida, Preco_Un, Total_Venda, Data_Hora = linha
                chave_vendas = novo_id_relatorio_vendas(vendas)

                vendas[chave_vendas] = {
                    "nome": Nome,
                    "preco": float(Preco_Un),
                    "quantidade_venda": int(Qtd_Vendida),
                    "total_venda": float(Total_Venda),
                    "data_hora": Data_Hora
                }

        print(f"Relatório de vendas importado com sucesso: {nome_arquivo}")

    except FileNotFoundError:
        print("Arquivo de relatório de vendas não encontrado. Nenhum dado foi carregado.")
    except Exception as e:
        print(f"Ocorreu um erro ao carregar as vendas anteriores: {e}")

# executa a função para importar as informações de vendas anteriores antes de inicializar o programa.
# dessa forma se garante a persistência de dados (atualização de estoque disponível)
carregar_vendas_anteriores()

menu_principal()
```
